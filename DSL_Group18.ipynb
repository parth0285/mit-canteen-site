```python
import kagglehub

path = kagglehub.dataset_download(
    "gunavenkatdoddi/eye-diseases-classification"
)

print(path)
```

```python
import os
os.listdir(path)
```

```python
dataset_path = os.path.join(path, "dataset")
os.listdir(dataset_path)
```

```python
import pandas as pd
import os

# Create an empty list to store image paths and labels
data = []

# Loop through each folder (each folder is a disease class)
for label in os.listdir(dataset_path):

    # Create full path of the class folder
    class_folder = os.path.join(dataset_path, label)

    # Check if it is actually a folder
    if os.path.isdir(class_folder):

        # Loop through each image inside the class folder
        for img in os.listdir(class_folder):

            # Store full image path and its label
            data.append([
                os.path.join(class_folder, img),  # image file path
                label                              # class name
            ])

# Convert the list into a Pandas DataFrame
df = pd.DataFrame(data, columns=["image_path", "label"])

# Display first 5 rows of the DataFrame
df.head()
```

```python
# Displays the first few image paths with their corresponding disease labels
```

```python
# Remove rows with missing values
df.dropna(inplace=True)

# Remove duplicate image entries
df.drop_duplicates(inplace=True)
# Removes missing and duplicate records from the dataset
```

```python
from PIL import Image
import numpy as np

def is_valid_image(path):
    try:
        img = Image.open(path)
        np.array(img)  # convert image to array
        return True
    except:
        return False

# Keep only valid images
df = df[df["image_path"].apply(is_valid_image)]
# Removes corrupt or unreadable image files
```
